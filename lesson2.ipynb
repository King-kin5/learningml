```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("dummy_expenses_30.csv")
## check for missing value
df.isnull().sum()
#drop rows with missing values
df.dropna(inplace=True)
#remove duplicates
df.drop_duplicates(inplace=True)
print(df)
```

```python
# 4. Extract a "year‑month" period from your dates
df['YearMonth'] = df['Date'].dt.to_period('M')  # :contentReference[oaicite:0]{index=0}

# 5. Group by that period and sum the amounts
monthly_sum = df.groupby('YearMonth')['Amount ($)'].sum().sort_index()  # :contentReference[oaicite:1]{index=1}

# 6. (For plotting) Convert the PeriodIndex back to actual timestamps
#monthly_sum.index = monthly_sum.index.to_timestamp()

# 7. Plot the aggregated series
monthly_sum.plot(kind='bar')  
plt.xlabel('Month')
plt.ylabel('Total Withdrawn ($)')
plt.title('Tracking Data by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
cat=df['Category'].value_counts()
cat.plot(kind='bar')
plt.title('')
plt.xlabel('category')
plt.ylabel('value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#This describes the  spots I had purchased most frequently. The data outputted is adjusted to protect my privacy
```
